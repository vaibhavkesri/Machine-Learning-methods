```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#loading data
dataset = pd.read_csv('SpamDetector/spam.csv',encoding='ISO-8859-1')
```

```python
dataset.head()
```

```python
#taking out first 2 columbs(relavent columbs)
dataset=dataset.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
dataset
```

```python
#making vocabulary using data set
vocabulary = {}
sentances = dataset['v2']
for sentance in sentances:
    for word in sentance.split():
        if word not in vocabulary:
            vocabulary[word]=0
```

```python
#Importinng libraries to convert sentances to vectors
from sklearn.feature_extraction.text import CountVectorizer
vectorizer =CountVectorizer()
```

```python
#Spliting data and label
#X = dataset['v2']
#y = dataset['v1']
```

```python
X = vectorizer.fit_transform(dataset['v2'])
```

```python
#Converting ham spam into '1' and '0'
def hamSpamConverter(s):
    if s == 'ham':
        return 1
    else:
        return 0
```

```python
y = dataset['v1'].apply(hamSpamConverter)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
print(X_train)
```

```python
#fitting weights of LinearRegression model according to train data
model.fit(X_train,y_train)
```

```python
#Predicting ham/spam for test sentances(messages)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
#Checking predicted ham/spam VS actual output from y_test
report = classification_report(y_test,y_pred)
```
